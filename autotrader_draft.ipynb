# 0.Project overview

**What is a derivative?**

Derivatives are a brief overview of details which are essential to car buyers. This will show in your advert title, along with your make and model.

Derivatives may include:

- Engine size: The engine size is measured in litres. The bigger the engine size, the more powerful the car is.
- Car body style: This depends on the size of your car and its different features. Examples include coupe, hatchback, convertible and estate.
- Number of Doors: The number of car doors, including the boot of your car. This can change depending on different car models.
- Number of Seats: The number of seats in your car when manufactured. This doesn't include any optional extras or modifications.
- Fuel type: This is the fuel you fill your car with. Examples are petrol, diesel, electric and hybrid cars (electric/petrol or electric/diesel)
- Transmission: The way your car is controlled and how you drive the car. Examples are manual and automatic.
- CO2 emissions: This is displayed as g/km. It shows how much carbon dioxide is released when driving the vehicle. The lower the car has, the better it is for the environment.
- BHP: Also known as brake horsepower. BHP is how much power your car releases and is measured at the flywheels. This shows the performance of the engine and how much work a car can do. The average car has around 100 BHP depending on engine size.
- Please note: PSI is different to BHP, PSI (pounds per square inch) is slightly different and is higher than your BHP

**Initial notes to consider:**

- Categorical variables like **make, model, fuel_type, transmission_type, and colour** may need to be one-hot encoded or converted into numerical values using other encoding techniques.
- Date fields like **first_date_seen and last_date_seen will** be used to calculate the **days_to_sell**, but won't be directly used as features unless you want to capture seasonality effects.
- Binary fields like **can_home_deliver and manufacturer_approved** can be used as-is or may be encoded as 0s and 1s.
- Continuous variables like **odometer_reading_miles and first_retailer_asking_price** can be used directly, though they may need scaling or normalization.
- Some features, like **stock_item_id and derivative_id**, are identifiers and should not be used as predictors since they do not hold any intrinsic predictive power.

Each of these features provides information that could potentialy be useful in building a predictive model for the "days_to_sell". To select the most relevant features for our model, we would typically perform feature selection techniques such as correlation analysis or using algorithms that include built-in feature selection like Lasso regression.

How does AutoTrader work out their "price_indicator_rating"?

They use data from millions of vehicles to work out a car’s current market price. This data includes things like:

- make
- Model
- Age
- Fuel type
- Optional extras and features

There are other things that are not included in how they work out the price labels, but these might affect the seller’s price:

- The car’s condition
- The car’s colour
- Supply and demand in the local area
- Special modifications (like a spoiler or body kit)
- Extra services (like warranties and finance deals)
- The number of owners
- The car’s service history

What if a car is spotted with no price label?

Not every car will have a price label, but there are lots of reasons why this might be. These reasons include if the car is:

- Brand new or nearly new
- A rare or classic car
- Being sold by a private seller
- Priced below £1,500 or above £70,000
- Over 15 years old
- Has been written off in the past (Category C/D/S/N)
- Has been imported from another country

```python
# simple func to show number of missing values
def show_missing_value_counts():
    return data_df.isna().sum()


def show_missing_value_counts_arg(df):
    return df.isna().sum()


# Function to get cardinality numbers for each categorical column of a given dataset.
def get_df_categorical_cols_cardinalities(df):
    return df.select_dtypes(include="object").nunique(dropna=False)


# Function to visualise unique value count proportions in a given column (Bar Chart)
def visualiseValueCountsProportions(df, colName, title=""):
    if title == "":
        title = colName
    fig, ax = plt.subplots()
    df[colName].value_counts(normalize=True, dropna=False).sort_values(
        ascending=True
    ).plot(kind="barh", ax=ax)
    ax.set_ylabel(colName)
    ax.set_facecolor("white")
    ax.set_xlabel("Proportion")
    ax.set_title(f"{title} Distribution")


# Same function but no chart.
def getValueCounts(df, colName):
    return df[colName].value_counts(dropna=False).sort_values(ascending=True)


# Function to visualise Avg DTS for each unique value in a given column (Bar Chart)
def visualiseValuesAvgDts(df, colName):
    ax = (
        df.groupby(colName)["days_to_sell"]
        .mean()
        .sort_values(ascending=True)
        .plot(kind="bar", ylabel="DTS", title=f"Avg DTS vs {colName}")
    )
def visualiseCorrelationUsingBarChart(df, srcCol, trgCol):
    
    srcDf = pd.get_dummies(data=df[[srcCol, trgCol]], dtype=int)
    fig, ax = plt.subplots()
    ax = srcDf.corr()[trgCol][1:].sort_values().plot(kind="bar")
    ax.set_title(f"{trgCol} Correlation with {srcCol}")



# func used to fill missing values on the "target_col" based on values obtained from cars with similiar "base_col" property
# for example: fill missing 0-60mph based on cars with the same derivative.
def fill_missing_vals(df,target_col_name,base_col_name):
    df_copy = df.copy()
    temp_dict = (
        df_copy[~df_copy[target_col_name].isna()]
        .set_index(base_col_name)[target_col_name]
        .to_dict()
    )
    df_copy[target_col_name] = df_copy[target_col_name].fillna(
        df_copy[base_col_name].map(temp_dict)
    )
    return df_copy

    

    
# func used to fill missing values on the "target_col" based on mean values when grouped by "base_col" values.
def fill_missing_vals_groupedby(df,target_col_name,base_col_name):
    df_copy = df.copy()
    temp_dict = (
        df_copy.groupby([base_col_name])
        .describe(exclude=["object", "datetime64"])
        .loc[:, (slice(None), ["mean"])][target_col_name]["mean"]
        .to_dict()
    )
    df_copy[target_col_name] = df_copy[
        target_col_name
    ].fillna(df_copy[base_col_name].map(temp_dict))

    return df_copy

```

Approaching imputation hierarchically can be an effective way to handle missing data, especially when there is a logical order or hierarchy to the data. In the context of our data, this makes sense because:

- **Derivative-Level Imputation:** Imputing at this level first is a good idea since it should provide the most specific and accurate imputation for vehicles within the same derivative category.

- **Make-Model-Year Imputation:** There are instances where derivative-level data is not available or there are still missing values, stepping back to the make, model, and year provides a broader category that still maintains some level of specificity. Different generations of the same model can have different performance characteristics, so this level of detail is valuable.

- **Make-Model Imputation:** Next, since there are still missing values after the first two steps, dropping back to the make and model level is broader still but can still offer a reasonable estimate for vehicles where more specific data points are not available.

- **Fuel_type:** Finally, for the rest of the missing values, we have used fuel type to impute mean values.

```python
data_df["zero_to_sixty_mph_seconds"].isna().sum()
```

```python
# The following code searches for the context in which the most common words from the attention_grabber field are used, 
# to validate the above insights:

# nltk.tokenize contains functions for tokenizing text into words and sentences.
from nltk.tokenize import word_tokenize, sent_tokenize
# nltk is the Natural Language Toolkit module.
import nltk

# Define a function to get concordance for a word (will find and display the context around a given word.)
# Parameters: the word to search for (word), the text to search within (text), the number of characters to
# show around the word (width), and the maximum number of lines to display (lines).
def get_concordance(word, text, width=40, lines=5):
    # Tokenize into sentences
    sentences = sent_tokenize(text)
    concordance_list = []
    
    # Search for the target word within sentences
    for sentence in sentences:
        if word in sentence:
            position = sentence.index(word)
            # Display a window of text around the word:
            # We use max to ensure the start index is not less than 0.
            start = max(0, position - width)
            end = position + width
            concordance_list.append(sentence[start:end])
            
            if lines and len(concordance_list) >= lines:
                break  # Stop if the specified number of lines is reached
    
    return concordance_list

# Print concordances for the top words
top_words = [word for word, frequency in fdist.most_common(20)]

# Combine all attention grabber text into one large string for analysis
full_text = ' '.join(data_df['attention_grabber'])

# Get concordances for each top word
for word in top_words:
    concordances = get_concordance(word, full_text)
    print(f"Concordances for '{word}':")
    for concordance in concordances:
        print(f"...{concordance}...")
    print('\n')  # Print newline for readability between words
```

```python
def remove_outliers_IQR(df:pd.DataFrame,column_name:str,lower_iqr_mult:float=1.5,upper_iqr_mult:float=1.5) -> pd.DataFrame:
    df_temp = df.copy(deep=True)

    Q1 = df_temp[column_name].quantile(0.25)
    Q3 = df_temp[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - lower_iqr_mult*IQR
    upper = Q3 + upper_iqr_mult*IQR

    print("Lower Bound And Higher Band : ",(lower,upper))
    
    df_filtered = df_temp[(df_temp[column_name] >= lower) & (df_temp[column_name] <= upper)]

    print("Number of Dropped Rows : "+str(len(df)-len(df_filtered)))
    return df_filtered

def remove_outliers_values(df:pd.DataFrame,column_name:str,lower:float,upper:float) -> pd.DataFrame:
    df_temp = df.copy(deep=True)

    print("Lower Bound And Higher Band : ",(lower,upper))
    
    df_filtered = df_temp[(df_temp[column_name] >= lower) & (df_temp[column_name] <= upper)]

    print("Number of Dropped Rows : "+str(len(df)-len(df_filtered)))
    return df_filtered
```

- **Gradient Boosting Machines (GBM):** Models such as XGBoost, LightGBM, or CatBoost can offer improved performance over Random Forests if tuned properly, but may require more computational resources and time.

**Questions:**

If our EDA suggests that relationships in the data are non-linear or that features interact in complex ways, we might need a model that can capture these complexities and Linear Regression might not be a good choice.
2. Which models can run in parallel?

   - Linear Regression:

     Some implementations of linear regression, particularly those that use gradient descent or coordinate descent, can be parallelized.

   - Ridge and Lasso Regression:

     These can also be parallelized, especially when using techniques like coordinate descent.

   - Elastic Net Regression:

     Similar to Ridge and Lasso, Elastic Net can be parallelized through certain solvers that support parallel computation.

   - Random Forest Regression:

     Random Forest inherently supports parallel processing because each tree in the forest can be built independently of the others.

   - Gradient Boosting Machines (GBM):

     Libraries like XGBoost, LightGBM, and H2O's GBM can utilize multi-core CPUs to build trees in parallel during the learning phase.

   - Support Vector Regression (SVR):

     Some implementations of SVR can be parallelized, but the parallelism might be limited compared to tree-based methods.

   - K-Nearest Neighbors (KNN):

     The process of finding the K-nearest neighbors can be parallelized, although the actual model is quite simple and has no training phase.

   - Stochastic Gradient Descent (SGD):

     This algorithm can run in parallel for large-scale problems using mini-batch training.

   - Decision Tree Regression:

     While a single decision tree is not parallelized, when using an ensemble of trees (like in a Random Forest), each tree can be built in parallel.

   - Neural Networks:

   Neural network training, particularly with libraries like TensorFlow and PyTorch, can be parallelized across multiple CPUs and GPUs.
   When leveraging parallel processing capabilities, you often need to explicitly enable this feature in the algorithm's implementation. For example, in the scikit-learn library, many estimators have an n_jobs parameter that allows you to specify the number of CPU cores to use for parallel processing (where n_jobs=-1 means using all available cores).

It's also important to note that while parallelism can significantly speed up computation, it may not always lead to faster convergence for certain algorithms, and the communication overhead between processes or threads can sometimes offset the performance gains. Additionally, not all steps in the training process of some algorithms can be parallelized, so the overall training time may not decrease linearly with the number of cores used.-
